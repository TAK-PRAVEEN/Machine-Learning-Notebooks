```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=1000, n_features=10, noise=0.1, random_state=1)
x, y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr_reg = LinearRegression()
dt_reg = DecisionTreeRegressor()
svm_reg = SVR()
```

```python
from sklearn.ensemble import VotingRegressor
ensemble_reg = VotingRegressor(estimators=[('Linear Regression',lr_reg), ('Decision Tree Regressor',dt_reg),('Support Vector Regressor',svm_reg)])
ensemble_reg
```

```python
ensemble_reg.fit(x_train, y_train)
```

```python
y_pred = ensemble_reg.predict(x_test)
y_pred
```

```python
ensemble_reg.named_estimators_
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
regressors = [lr_reg, dt_reg, svm_reg]
for reg in regressors:
    reg.fit(x_train, y_train)
    y_pred = reg.predict(x_test)
    print(f"Accuracy score by {reg} is {r2_score(y_test, y_pred)}")
    print("================================================================\n")
```
