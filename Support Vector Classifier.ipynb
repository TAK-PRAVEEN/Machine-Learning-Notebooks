```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0, random_state=1)
x, y
```

```python
pd.DataFrame(x)[0]
```

```python
pd.DataFrame(x)[1]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, y_train.shape
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred) 
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'C':[9,9.1,9.2,9.3,8.8,8.9,8.99],
    'gamma':[0.2,0.1,0.11,0.15,0.21,0.25,0.15,0.19],
    'kernel':['linear']
}
```

```python
model = GridSearchCV(estimator=classifier, param_grid=params, cv=5, verbose=3)
model
```

```python
model.fit(x_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
accuracy_score(y_test, y_pred) #0.8566666666666667
```

```python
best_model = model.best_estimator_
best_model
```

```python
y_pred_new = best_model.predict(x_test)
y_pred_new
```

```python
accuracy_score(y_test, y_pred_new) #0.8566666666666667
```
