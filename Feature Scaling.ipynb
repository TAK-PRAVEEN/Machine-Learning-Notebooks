```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Scaling(optional)>>brings all features on same scale
```

```python
#Why Scaling?>> To reduce the computation time (faster optimization) and better interpretation
```

```python
df=sns.load_dataset('tips')
df
```

```python
#Standardization>> ascore=x-mu/sigma
```

```python
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
```

```python
std_data=[]
for i in list(df['total_bill']):
    std_data.append((i-mean)/std)
```

```python
std_data
```

```python
#Scaling is optinal because it doesn't change the distribution of data
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
df
```

```python
scaler.fit_transform(df[['total_bill']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tips'])
```

```python
scaler.transform([[13,1]])
```

```python
#Normalization/Min Max scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
```

```python
data=min_max.fit_transform(df[['total_bill','tip']])
data
```

```python
sns.distplot(data)
```

```python
scaled=[]
for i in list(df['total_bill']):
    scaled.append((i-min(df['total_bill']))/(max(df['total_bill'])-min(df['total_bill'])))
```
