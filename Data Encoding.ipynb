```python
# Target Guided Ordinal Encoding
# Relation between target variables
# A lot of unique categories
# Replace with mean or median
```

```python
df=pd.DataFrame({'time':['lunch','breakfast','breakfast','dinner','lunch','breakfast','dinner','breakfast','dinner','lunch'],'total_bill':[120,150,200,130,180,200,250,180,160,120]})
df
```

```python
mean_price=dict(df.groupby('time')['total_bill'].mean())
mean_price
```

```python
df['time_target_encoded']=df['time'].map(mean_price)
df
```
