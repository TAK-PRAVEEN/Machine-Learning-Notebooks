```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=1)
x, y # No need of y in unsupervised learning
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, x_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans
```

```python
kmeans.fit(x_train)
```

```python
kmeans.labels_
```

```python
y_labels = kmeans.predict(x_train)
y_labels
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=y_labels)
```

```python
# how better the clustering model is
from sklearn.metrics import silhouette_score
silhouette_score(x_train, kmeans.labels_)
```

```python
kmeans.inertia_ # for k = 3, you have WCSS distance
```

```python
# To select K (centroid) >> elbow method
WCSS = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_train)
    WCSS.append(kmeans.inertia_)
print(WCSS)
```

```python
#Plot on elbow curve
plt.plot(range(1,10), WCSS)
plt.xticks(range(1,10))
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.show() # 3 is constant
# Based on plot the optimal number of k cluster is 3
```

```python
from kneed import KneeLocator
k1 = KneeLocator(range(1,10), WCSS, curve='convex', direction='decreasing')
k1
```

```python
k1.elbow
```

```python
silhouette_coeff = []
for k in range(2,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_train)
    score = silhouette_score(x_train, kmeans.labels_)
    silhouette_coeff.append(score)
print(silhouette_coeff)
```
