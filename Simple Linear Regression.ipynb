```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#read the data
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes
```

```python
data=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
data
```

```python
data['target']=diabetes.target
```

```python
data
```

```python
#data preparation
#divide the data into x,y
#Simple Linear Regression
x=data[['bmi']]
y=data[['target']]
```

```python
x,y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=111)
```

```python
x_train
```

```python
y_train
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
#model training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
print(f"Shape of X train: {x_train.shape}")
print(f"Shape of y train: {y_train.shape}")
print(f"Shape of X test: {x_test.shape}")
print(f"Shape of y test: {y_test.shape}")
print(f"Shape of y pred: {y_pred.shape}")
```

```python
sns.scatterplot(x=data['bmi'],y=data['target'],hue=data['bmi'])
```

```python
#Visualize the result
plt.scatter(x_test,y_test,color="black",label="Actual Data")
plt.plot(x_test,y_pred,color="red",linewidth=3,label="linear rregression line")
plt.xlabel("BMI")
plt.ylabel("one year Progression- target")
plt.title("Linear Regression Model")
plt.legend()
plt.show()
```
