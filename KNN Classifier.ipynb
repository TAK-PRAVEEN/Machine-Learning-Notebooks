```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=3, n_redundant=0, n_classes=2, random_state=1)
x, y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
clf
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors' : [3,5,7,1,10,20,25,30],
    'algorithm' : ['ball_tree','kd_tree','brute'],
    'leaf_size' : [20,30,40,15]
}
```

```python
grid = GridSearchCV(estimator=clf, param_grid=params, cv=5, verbose=3)
grid
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)
accuracy_score(y_test, y_pred)
```
