DATA INTERPOLATION

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,5,6,8])
plt.scatter(x,y,c='r')
sns.lineplot(x=x,y=y,color='orange')
```

```python
#Linear Interpolation
x_new=np.linspace(1,5,10)
y_interp=np.interp(x_new,x,y)
```

```python
x_new,y_interp
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x_new,y_interp)
sns.lineplot(x=x_new,y=y_interp,color='blue')
```

```python
#Cubic Interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
plt.scatter(x,y,c='r')
sns.lineplot(x=x,y=y,color='orange')
```

```python
from scipy.interpolate import interp1d
f=interp1d(x,y,kind="cubic")
f
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
```

```python
x_new,y_interp
```

```python
plt.scatter(x_new,y_interp,c='r')
sns.lineplot(x=x_new,y=y_interp,color='orange')
```

```python
#Polynomial Interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,1,25])
plt.scatter(x,y)
sns.lineplot(x=x,y=y)
```

```python
p=np.polyfit(x,y,2)
p
```

```python
x
```

```python
x_new=np.linspace(1,5,10)
y_interp=np.polyval(p,x_new)
plt.scatter(x_new,y_interp)
sns.lineplot(x=x_new,y=y_interp)
```

OUTLIER TREATMENT

```python
salary=[11,40,45,68,65,68,78,90,57,74,91,92,88,68,57,48,99,101,68,77,110,140]
```

```python
#Five point summary
np.quantile(salary,[0,0.25,0.50,0.75,1])
```

```python
df=pd.DataFrame(salary,columns=['Salary'])
df
```

```python
df.shape
```

```python
#To check outliers>> displot, boxplot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df.Salary,kde=True)
plt.title("Dist Plot")

plt.subplot(1,2,2)
sns.boxplot(data=df,x='Salary')
plt.title("Box Plot")
plt.show()
```

```python
#Drop the outliers
# Caping the outliers
# Replace with mean or median
# Scaling and transformation
```

```python
q1=df.Salary.quantile(0.25)
q3=df.Salary.quantile(0.75)
```

```python
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
lower_fence=q1-1.5*IQR
upper_fence=q3+1.5*IQR
```

```python
lower_fence,upper_fence
```

```python
#Dropppinf the outliers
df.shape
```

```python
df_filtered=df[(df.Salary>=lower_fence) & (df.Salary<=upper_fence)]
df_filtered
```

```python
df_filtered.shape
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df_filtered['Salary'],kde=True)
plt.title("Dist Plot")

plt.subplot(1,2,2)
sns.boxplot(data=df_filtered,x="Salary")
plt.title("Box Plot")
plt.show()
```

```python
#Replace the ouliers with mean and median
df['Salary_mean']=np.where((df.Salary>=upper_fence) | (df.Salary<=lower_fence),df.Salary.mean(),df.Salary)
```

```python
df.Salary.mean()
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df['Salary_mean'],kde=True)
plt.title("Dist Plot")

plt.subplot(1,2,2)
sns.boxplot(data=df,x="Salary_mean")
plt.title("Box Plot")
plt.show()
```

```python
#Capping>> Replacing ouliers with the nearest value i.e. not outlier
lower_cap=df.Salary.quantile(0.05)
upper_cap=df.Salary.quantile(0.95)
```

```python
lower_cap,upper_cap
```

```python
df['Salary_cap']=np.where((df.Salary<lower_cap),lower_cap,np.where((df.Salary>upper_cap),upper_cap,df.Salary))
```

```python
df
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df['Salary_cap'],kde=True)
plt.title("Dist Plot")

plt.subplot(1,2,2)
sns.boxplot(data=df,x="Salary_cap")
plt.title("Box Plot")
plt.show()
```
