```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('mpg')
df
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
df.isna().sum()
```

```python
df.info()
```

```python
df['origin'].unique()
```

```python
df['origin']=df['origin'].map({"usa":1,"japan":2,"europe":3})
df.info()
```

```python
df.info()
```

```python
x=df.drop('mpg',axis=1)
y=df['mpg']
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
for i,columns in enumerate(x_train.columns):
  print(f'The coefficient for {columns} is {model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#RIDGE REGRESSION
```

```python
from sklearn.linear_model import Ridge
ridge_regression_model=Ridge(alpha=0.1)
ridge_regression_model.fit(x_train,y_train)
for i,column in enumerate(x_train.columns):
  print(f'The coefficient for {column} is {ridge_regression_model.coef_[i]}')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred) #0.8348001123742284
```

```python
#LASSO REGRESSION
```

```python
from sklearn.linear_model import Lasso
lasso_regression_model=Lasso(alpha=0.1)
lasso_regression_model.fit(x_train,y_train)
for i,column in enumerate(x_train.columns):
  print(f"The coefficient for {column} is {lasso_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred) #0.8348001123742284
```

```python
#ELASTIC NET
```

```python
from sklearn.linear_model import ElasticNet
elasticnet_regression_model=ElasticNet(alpha=0.1)
elasticnet_regression_model.fit(x_train,y_train)
for i,column in enumerate(x_train.columns):
  print(f"The coefficient for {column} is {elasticnet_regression_model.coef_[i]}")
```
