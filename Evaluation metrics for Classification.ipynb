```python
thresholds = np.linspace(0,1,100)
thresholds
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precisons = []
recalls = []
accuracies = []
for threshold in thresholds:
    y_pred_threshold = (y_pred_prob >= threshold).astype(int)
    print(y_pred_threshold)
    precison = precision_score(y_test,y_pred_threshold)
    recall = recall_score(y_test,y_pred_threshold)
    accuracy = accuracy_score(y_test,y_pred_threshold)
    precisons.append(precison)
    recalls.append(recall)
    accuracies.append(accuracy)
```

```python
plt.figure(figsize=(10,6))
plt.plot(thresholds, precisons, label="Precisions")
plt.plot(thresholds, recalls, label="Recalls")
plt.plot(thresholds, accuracies, label="Accuracies")
plt.xlabel("Threshold Probability")
plt.ylabel("Score")
plt.title("Precision, Recall, and Accuracy vs, Threshold Probability")
plt.legend()
plt.grid()
plt.show()
```
